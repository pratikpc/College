```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/linear-reg-least-squares.csv')
X = data.iloc[:, 0]/70
Y = data.iloc[:, 1]/120
```

```python
print(data.head())
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
meanX = np.mean(X)
meanY = np.mean(Y)
print(meanX)
```

```python
num = sum([(x-meanX)*(y-meanY) for (x,y) in zip(X,Y)])
den = sum((x-meanX)**2 for x in X)
slope = num/den
```

```python
const = meanY - slope*meanX
```

```python
print (slope, const)
```

```python
A = slope*X + const
```

```python
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(A), max(A)], color='red') # predicted
plt.show()
```
